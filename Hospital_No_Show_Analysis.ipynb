Analyze hospital appointment data to identify patterns, trends, and business insights related to patient no-shows.

- Matplotlib  
- Seaborn  

---


```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```

df = pd.read_csv('noshowappointments.csv')
df.head()
```

## Dataset Overview

```

df.shape
```

```

df.columns
```

## Data Cleaning

```

# Check missing values
df.isnull().sum()
```

```

# Check duplicates
df.duplicated().sum()
```

```

# Remove duplicates
df = df.drop_duplicates()
```

```

# Rename column for easier use
df.rename(columns={'No-show':'No_show'}, inplace=True)
```

## Descriptive Analysis

```

# Total records
len(df)
```

```

# Appointment status count
df['No_show'].value_counts()
```

```

# Percentage distribution
df['No_show'].value_counts(normalize=True)*100
```

```

# Neighbourhood counts
df['Neighbourhood'].value_counts().head(10)
```

## Visualizations

```

# Appointment Status Chart
df['No_show'].value_counts().plot(kind='bar')
plt.title("Appointment Status")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

```

# Top 10 Neighbourhoods Chart
df['Neighbourhood'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Neighbourhoods")
plt.xlabel("Neighbourhood")
plt.ylabel("Count")
plt.show()
```

## Trend Analysis

```

df.groupby('Neighbourhood')['No_show'].value_counts().unstack()
```


## Observations

1. Total 110,527 appointments were analyzed.

2. Around 20% patients missed their appointments.

3. Majority of patients attended their appointments.

4. Certain neighbourhoods have higher no-show rates.

5. No missing values were found in dataset.

---

## Business Recommendation

Hospitals can send reminders and focus on neighbourhoods with high no-show rates to reduce missed appointments.
